```python
import pandas as pd
import plotly.express as px
import streamlit as st

st.set_page_config(page_title="IBA - Purchase",
                   page_icon=":bar_chart:",
                   layout="wide"
)

def get_data_from_excel():
    df = pd.read_excel(
        io='Purchase orders_637997872068804066.xlsx',
        engine='openpyxl',
        # sheet_name='Processed',
        skiprows=0,
        # usecols='B:AI',
        # nrows=789
    )

    # df.columns = df.columns.str.replace(' ', '')

    return df

df = get_data_from_excel()
```

```python
approval_status_selector = st.multiselect(
    "Approval Status:",
    options=df['Approval status'].unique(),
    default=df['Approval status'].unique()
)

df_selection = df.query('`Approval status` == @approval_status_selector')
```
